```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-40", "40+"]
age_counts = [17, 22, 107, 258, 77, 52, 31, 12]
colors = ["yellow","green","lightblue","orange","red","purple","yellowgreen","lightskyblue"]
explode = (0,0,0,0.1,0,0,0,0)
```

```python
plt.pie(age_counts, explode=explode, labels=ages, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=60)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
```

```python
plt.title("Players Shown By Age group")
```
